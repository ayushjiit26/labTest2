```python
from sklearn.datasets import load_diabetes
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = load_diabetes()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target, name='target')
X.hist(figsize=(15, 10), edgecolor='white')
plt.suptitle("Distribution of Each Feature")
plt.show()
num_features = X.select_dtypes(include=[np.number]).shape[1]
cat_features = X.select_dtypes(exclude=[np.number]).shape[1]
num_classes = len(y.unique())
print(f"Number of numerical attributes: {num_features}")
print(f"Number of categorical attributes: {cat_features}")
print(f"Number of classes in dataset: {num_classes}")
```

```python
print(X.corr())
sns.heatmap(X.corr(),annot=True, cmap='coolwarm')
plt.title("Heatmap of Feature Correlations")
plt.show()
print("END")
```
